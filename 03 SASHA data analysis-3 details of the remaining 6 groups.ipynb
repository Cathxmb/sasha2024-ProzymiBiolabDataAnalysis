```python
import sklearn
from sklearn import datasets 

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# to remove the "%" and remain the same value of the "weight"
food_products = pd.read_csv('jp_lca_dat.csv')
food_products
```

```python
# to obtain the data
food_fruit_group = food_products[food_products["Group"] == "fruit"]
food_vegetables_group = food_products[food_products["Group"] == "vegetables"]
food_sugars_group = food_products[food_products["Group"] == "sugars"]
food_alcohols_group = food_products[food_products["Group"] == "alcohols"]
# starch_rich
food_starch_rich_group = food_products[food_products["Group"] == "starch_rich"]
# milks
food_milks_group = food_products[food_products["Group"] == "milks"]
# oils
```

```python
# to display the details of the fruit
food_fruit_group
```

```python
food_fruit_group_details = food_fruit_group["Product_details"].unique()
food_fruit_group_details
```

```python
print(food_fruit_group.shape)
```

```python
#Land use
# create a grid of the plots 
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(25, 18))
for ax in [ax1, ax2, ax3, ax4, ax5, ax6]:
    ax.tick_params(axis='x', rotation=30)
    
ax1.scatter(food_vegetables_group["Product_details"], food_vegetables_group["Land Use (m2*year)"], c="brown")
ax1.set_title('Vegetables')

ax2.scatter(food_fruit_group["Product_details"], food_fruit_group["Land Use (m2*year)"], c="yellow")
ax2.set_title('Fruit')

ax3.scatter(food_sugars_group["Product_details"], food_sugars_group["Land Use (m2*year)"], c="blue")
ax3.set_title('Sugars')

ax4.scatter(food_alcohols_group["Product_details"], food_alcohols_group["Land Use (m2*year)"], c="grey")
ax4.set_title('Alcohols')

ax5.scatter(food_milks_group["Product_details"], food_milks_group["Land Use (m2*year)"], c="purple")
ax5.set_title('Milks')

ax6.scatter(food_starch_rich_group["Product_details"], food_starch_rich_group["Land Use (m2*year)"], c="orange")
ax6.set_title('Starch rich')

# plt.xticks(rotation=30, ha='right')

# the space between subplots
# plt.subplots_adjust(hspace=0.8)
plt.tight_layout()
plt.show()
```

```python
# Water use
# create a grid of the plots
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(25, 18))
for ax in [ax1, ax2, ax3, ax4, ax5, ax6]:
    ax.tick_params(axis='x', rotation=30)
    
ax1.scatter(food_vegetables_group["Product_details"], food_vegetables_group["Water Use (L)"], c="brown")
ax1.set_title('Vegetables')

ax2.scatter(food_fruit_group["Product_details"], food_fruit_group["Water Use (L)"], c="yellow")
ax2.set_title('Fruit')

ax3.scatter(food_sugars_group["Product_details"], food_sugars_group["Water Use (L)"], c="blue")
ax3.set_title('Sugars')

ax4.scatter(food_alcohols_group["Product_details"], food_alcohols_group["Water Use (L)"], c="grey")
ax4.set_title('Alcohols')

ax5.scatter(food_milks_group["Product_details"], food_milks_group["Water Use (L)"], c="purple")
ax5.set_title('Milks')

ax6.scatter(food_starch_rich_group["Product_details"], food_starch_rich_group["Water Use (L)"], c="orange")
ax6.set_title('Starch rich') 

# the space between subplots
# plt.subplots_adjust(wspace=0.5, hspace=0.8)
plt.tight_layout()
plt.show()
```

```python
# GHG Emissions 
# create a grid of the plots
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(25, 18))
for ax in [ax1, ax2, ax3, ax4, ax5, ax6]:
    ax.tick_params(axis='x', rotation=30)
    
ax1.scatter(food_vegetables_group["Product_details"], food_vegetables_group["GHG Emissions (kg CO2eq, IPCC2013 incl CC feedbacks)"], c="brown")
ax1.set_title('Vegetables')

ax2.scatter(food_fruit_group["Product_details"], food_fruit_group["GHG Emissions (kg CO2eq, IPCC2013 incl CC feedbacks)"], c="yellow")
ax2.set_title('Fruit')

ax3.scatter(food_sugars_group["Product_details"], food_sugars_group["GHG Emissions (kg CO2eq, IPCC2013 incl CC feedbacks)"], c="blue")
ax3.set_title('Sugars')

ax4.scatter(food_alcohols_group["Product_details"], food_alcohols_group["GHG Emissions (kg CO2eq, IPCC2013 incl CC feedbacks)"], c="grey")
ax4.set_title('Alcohols')

ax5.scatter(food_milks_group["Product_details"], food_milks_group["GHG Emissions (kg CO2eq, IPCC2013 incl CC feedbacks)"], c="purple")
ax5.set_title('Milks')

ax6.scatter(food_starch_rich_group["Product_details"], food_starch_rich_group["GHG Emissions (kg CO2eq, IPCC2013 incl CC feedbacks)"], c="orange")
ax6.set_title('Starch rich')

# the space between subplots
# plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.tight_layout()
plt.show()
```

```python
# Biodiversity 
# create a grid of the plots
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(25, 18))
for ax in [ax1, ax2, ax3, ax4, ax5, ax6]:
    ax.tick_params(axis='x', rotation=30)

ax1.scatter(food_vegetables_group["Product_details"], food_vegetables_group["Biodiversity (sp*yr*10^14)"], c="brown")
ax1.set_title('Vegetables')

ax2.scatter(food_fruit_group["Product_details"], food_fruit_group["Biodiversity (sp*yr*10^14)"], c="yellow")
ax2.set_title('Fruit')

ax3.scatter(food_sugars_group["Product_details"], food_sugars_group["Biodiversity (sp*yr*10^14)"], c="blue")
ax3.set_title('Sugars')

ax4.scatter(food_alcohols_group["Product_details"], food_alcohols_group["Biodiversity (sp*yr*10^14)"], c="grey")
ax4.set_title('Alcohols')
# ax4.set_xticks(ax4.get_xticks(), rotation=45)

ax5.scatter(food_milks_group["Product_details"], food_milks_group["Biodiversity (sp*yr*10^14)"], c="purple")
ax5.set_title('Milks')
# ax5.set_xticks(ax5.get_xticks(), rotation=45)

ax6.scatter(food_starch_rich_group["Product_details"], food_starch_rich_group["Biodiversity (sp*yr*10^14)"], c="orange")
ax6.set_title('Starch rich')
# ax6.set_xticks(ax6.get_xticks(), rotation=45)

plt.xticks(rotation=30, ha='right')

# the space between subplots
# plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.tight_layout()
plt.show()
```
